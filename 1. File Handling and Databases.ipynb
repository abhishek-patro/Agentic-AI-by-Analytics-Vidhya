```python
import json
import pandas as pd
```

#### Reading the JSON data from a JSON file

```python
with open("1.a. JSON dumps/data.json") as file:
    data = json.load(file)
print(data)
```

#### Writing JSON data into a `.JSON` file

```python
with open("data.json",'w') as file:
    json.dump(data,file)
```

#### Use indent parameter in json dump to make the data more readable

```python
with open("data.json",'w') as file:
    json.dump(data,file, indent = 4)

print(data)
```

#### Since we have the data in a variable how to put it in a more readable format?

```python
orders = data["orders"]

for order in orders:
    print("Order ID:", order["order_id"])
    print("Customer:", order["customer"]["name"])

    total = 0
    for item in order["items"]:
        total += item["price"] * item["quantity"]

    print("Total Price:", total)
    print("Status:", order["status"])
    print("-" * 40)
```

#### How do we put the data in a tabular format? --> `pd.json_normalize`

```python
pd.json_normalize(data['orders'],record_path="items" ,\
                  meta = [
                            'order_id', \
                            ["customer","name"], ["customer","city"],\
#                             ["items","product"], ["items","price"], ["items","quantity"],\
                            ["status"]
                         ]

                 )
```

```python
pd.json_normalize(
    data["orders"],
    record_path="items",
    meta=["order_id", ["customer", "name"], ["customer", "city"], "status"]
)
```

Here is that concept formatted in clean, scannable Markdown:

---
JSON dumps get truly "complex" when they contain lists of objects *inside* the main objects. To flatten these, you need to explicitly tell `json_normalize()` two things:

* **What to unpack:** The specific nested list you want to expand (using the `record_path` argument).
* **What to keep:** The parent fields you want to retain alongside that unpacked data (using the `meta` argument).
