## 1. Simple Route to Greet User 

```python
from flask import Flask, request, render_template_string
```

```python
app = Flask(__name__)

@app.route('/')
def greet():
    return "Hello World! User🫂"

if __name__ == '__main__':
    app.run(debug = False)
```

```python
app = Flask(__name__)

@app.route('/')
def home():
    return "Hello World! User🫂"

@app.route('/greet/<name_input>')
def greet(name_input):
    return render_template_string(r'<h1>Hello, {{name}}</h1>',name = name_input)

@app.route('/square/<int:number>')
def square(number):
    return f"Your number is {number**2}!"

if __name__ == '__main__':
    app.run(debug = False)

# greet url - http://127.0.0.1:5000/greet/ahi
# square - http://127.0.0.1:5000/square/10

# Url encodings - if there are any special character in the url we need to enter it accordign to the url encodings set

# For eg: If we want to give a space, we have to write "John%20Miller"
# http://127.0.0.1:5000/greet/John%20Miller
```
